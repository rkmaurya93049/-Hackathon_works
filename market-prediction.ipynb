```python
feature_groups = {
    'D': [col for col in df.columns if col.startswith('D')],
    'E': [col for col in df.columns if col.startswith('E')],
    'I': [col for col in df.columns if col.startswith('I')],
    'M': [col for col in df.columns if col.startswith('M')],
    'P': [col for col in df.columns if col.startswith('P')],
    'S': [col for col in df.columns if col.startswith('S')],
    'V': [col for col in df.columns if col.startswith('V')],
}
```

```python
df.head()
```

```python
for group, cols in feature_groups.items():
    df[cols] = df[cols].fillna(df[cols].median())
```

```python
df.isnull().sum()
```

```python
for col in ['forward_returns', 'market_forward_excess_returns']:
    df[f'{col}_lag1'] = df[col].shift(1)
    df[f'{col}_rolling_mean5'] = df[col].rolling(window=5).mean()
```

```python
target = 'market_forward_excess_returns'
features = sum(feature_groups.values(), []) + ['forward_returns_lag1', 'market_forward_excess_returns_rolling_mean5']
df_model = df.dropna(subset=features + [target])
```

```python
train = df_model[df_model['date_id'] < df_model['date_id'].max() - 180]
test = df_model[df_model['date_id'] >= df_model['date_id'].max() - 180]
```

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor()
model.fit(train[features], train[target])
preds = model.predict(test[features])
```

```python
def map_to_allocation(pred, threshold=0.0):
    return np.clip(1 + pred, 0, 2)

allocations = map_to_allocation(preds)
```

```python
# Simulated daily portfolio returns
simulated_returns = allocations * test['forward_returns']

# Calculate cumulative return
cumulative_return = (1 + simulated_returns).cumprod()

# Estimate volatility
volatility = simulated_returns.std()

# Estimate Sharpe-like score (simplified)
excess_returns = simulated_returns - test['risk_free_rate']
sharpe_like = excess_returns.mean() / excess_returns.std()
```

```python
df_test = pd.read_csv('/kaggle/input/hull-tactical-market-prediction/test.csv')

# Apply same feature grouping
feature_groups = {
    'D': [col for col in df_test.columns if col.startswith('D')],
    'E': [col for col in df_test.columns if col.startswith('E')],
    'I': [col for col in df_test.columns if col.startswith('I')],
    'M': [col for col in df_test.columns if col.startswith('M')],
    'P': [col for col in df_test.columns if col.startswith('P')],
    'S': [col for col in df_test.columns if col.startswith('S')],
    'V': [col for col in df_test.columns if col.startswith('V')],
}

for group, cols in feature_groups.items():
    df_test[cols] = df_test[cols].fillna(df_test[cols].median())
```

```python
df_test['lagged_forward_returns'] = df_test['lagged_forward_returns'].fillna(0)
df_test['lagged_market_forward_excess_returns'] = df_test['lagged_market_forward_excess_returns'].fillna(0)
```

```python
features = sum(feature_groups.values(), []) + ['lagged_forward_returns', 'lagged_market_forward_excess_returns']
```

```python
preds_test = model.predict(df_test[features])
```

```python
def map_to_allocation(pred, threshold=0.0):
    return np.clip(1 + pred, 0, 2)

allocations_test = map_to_allocation(preds_test)
```

```python
submission = pd.DataFrame({
    'date_id': df_test['date_id'],
    'allocation': allocations_test
})
```
